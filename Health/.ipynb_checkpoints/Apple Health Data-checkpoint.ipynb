```python
import xml.etree.ElementTree as ET
from lxml import etree
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
xml_data = '/Users/ryand/Downloads/apple_health_export/export.xml'
```

```python
tree = etree.parse(xml_data)
root = tree.getroot()
```

```python
records = []
for i, element in enumerate(root):
    records.append(element.attrib)

HR_data = []
for i, record in enumerate(records):
    try:
        if record['type'] == 'HKQuantityTypeIdentifierHeartRate':
            HR_data.append(record)
    except:
        pass
HR_data_ = [dict(record) for record in HR_data]

df = pd.DataFrame(HR_data_, columns = HR_data[0].keys())
```

```python
time_col = ['startDate','endDate']
for col in time_col:
    df[col] = pd.to_datetime(df[col])
df['value'] = df['value'].astype(float)
```

```python
df = df.sort_values('startDate')
```

```python
df['hour'] = df['startDate'].map(lambda x: x.hour)
df['weekday'] = df['startDate'].map(lambda x: x.dayofweek)
```

```python
# df_ = df[df['startDate'] > '2017-05-15 00:00:00']
df_ = df.groupby(['weekday','hour']).mean() # Monday is 0
df_.index = np.arange(168)

fig, ax = plt.subplots(figsize=(15,9))
ax.plot(df_.index, df_['value'])
```
